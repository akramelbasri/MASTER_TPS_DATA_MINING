## <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;padding:5px;">Dropping columns that are of no use</p>


### Suppression selon le nombre des valeurs nulles : 


- Les colonnes contenant plus de 70 % de valeurs manquantes seront supprimées.

```python
columns_to_drop = []
for column in boston.columns:
    null_count = boston[column].isna().sum()
    total_rows = boston.shape[0]
    if (null_count / total_rows) >= 0.7:
        columns_to_drop.append(column)
        print(f"{column}: {null_count}")
```

```python
boston.drop(columns=columns_to_drop, axis=1, inplace=True)
print(f" - Le nombre de colonnes supprimées est : {len(columns_to_drop)}")
print(" - Les noms des colonnes supprimées : \n", columns_to_drop )
print("\n - Le nombre des colonnes restent :", len(boston.columns))
```

### - Suppression selon le nombre des enregitrements dupliquées :

```python
nbr_duplicate = boston.duplicated().sum()
print("- Le nombre des doublons dans notre dataset est: ", nbr_duplicate)
print("- Le nombre des colonnes restent :", len(boston.columns))
```

### - Suppression des villes qui n'appartiennent pas à Boston

```python
cities_to_remove = ['Brookline', 'Somerville', 
                    '88 Auckland Street, Apt 1, Dorchester, MA 0212', 
                    'Everett', 'Cambridge']

boston = boston[~boston['city'].isin(cities_to_remove)]
boston["city"].unique()
```

- Encoder la colonne "city"

```python
label_encoder = LabelEncoder()
boston['city_encoded'] = label_encoder.fit_transform(boston['city'])
boston = boston.drop(columns=['city'])
boston['city_encoded'].head(3)
```

### Suppression selon le nombre des valeurs uniques:

- Au sein de cette section, nous avons exclu toutes les colonnes ne contenant qu'une seule valeur unique répétée dans la totalité de la colonne.

On observe que le colonne "experiences_offered" ne contient qu'une des valeurs none, Il faut le supprimer aussi


```python
#Les valeurs uniques du colonne "experiences_offered"
boston["experiences_offered"].unique()
```

```python
boston = boston.drop(columns=["experiences_offered"], axis=1)
```

```python
# Pour chaque colonne en affiche les noms des colonnes ayant un nombre des valeurs uniques qui égale a 1 et le supprimer
cols_to_remove = [col for col in boston.columns if len(boston[col].unique()) == 1]
print(f"-Les colonnes a une seule valeur unique:\n {cols_to_remove}\n-total des colonnes supprimées: {len(cols_to_remove)}")
# Supprimer les colonnes du DataFrame
boston = boston.drop(cols_to_remove, axis=1)
```

```python
print("- Le nombre des colonnes restent :", len(boston.columns))
```

### - Suppression des identifiants et des données ont une relation avec celles récupérées

```python
boston.drop(columns=['id', 'last_scraped', 'host_id'], axis=1, inplace=True)
```

### Suppression selon le type:

- Dans cette section, nous allons lister le type de chaque colonne.
Pour les colonnes textuelles qui ne présentent aucune modalité, elles seront supprimées.
- Si une colonne textuelle contient des catégories spécifiques inferieurs de 10 modalitées , elles seront représentées numériquement.
- En ce qui concerne les colonnes numériques, nous examinerons leurs corrélations et supprimerons l'une d'entre elles en cas de corrélation élevée entre elles.

```python
# les type de chaque colonne
column_types = boston.dtypes
columns_to_drop = []
# Suppression des colonnes textuelles avec plus de 10 modalités
text_columns = column_types[column_types == 'object'].index
for column in text_columns:
    if len(boston[column].unique()) >= 10:
        boston = boston.drop(columns=[column])
        columns_to_drop.append(column)
print(f"- Le nombre des colonnes textuelles supprimées est: {len(columns_to_drop)},\n- Les noms des colonnes: {columns_to_drop} ")
print("- Le nombre des colonnes restent :", len(boston.columns))
```

## <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;padding:5px;border-radius:10px 10px;">Data Labelling</p>


- Représentation numérique des catégories des colonnes textuelles ayant des modalités limitées

```python
col_to_numeriques=[]
for column in text_columns:
    if column in boston.columns and len(boston[column].unique()) > 1:
        boston[column] = pd.Categorical(boston[column]).codes
        col_to_numeriques.append(column)
```

```python
print(f"Les colonnes textuelles qu'on a encoder en format numériques sont:\n {col_to_numeriques}")
boston.head(2)
```

## <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;padding:5px">Data Transform</p>


### - Nous allons transférer 't' à 1 et 'f' à 0 dans ces colonnes

```python
columns_to_transform = ['host_is_superhost', 'host_has_profile_pic', 'host_identity_verified', 'is_location_exact', 
                        'has_availability', 'requires_license', 'instant_bookable', 'is_business_travel_ready', 
                        'require_guest_profile_picture', 'require_guest_phone_verification']
mapping = {'t': 1, 'f': 0, 'nan': 0, np.nan: 0}

existing_columns = [col for col in columns_to_transform if col in boston.columns]

boston[existing_columns] = boston[existing_columns].replace(mapping)
```

```python
boston.head(3)
```

### - Convertir la colonne 'prix' 'security_deposit' 'cleaning_fee' 'extra_people' en float

```python
boston.columns = boston.columns.str.strip()
columns_to_transform = ['price', 'security_deposit', 'cleaning_fee', 'extra_people']

for col in columns_to_transform:
    if col in boston.columns:
        if boston[col].dtype == 'object':
            boston[col] = boston[col].str.replace('[\$,]', '', regex=True)
            # Convert values to float
            boston[col] = boston[col].astype(float)
```

```python
boston.head(3)
```

## <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;padding:5px;">Filling missing values</p>


### La Remplissage des colonnes ayant des valeurs manquantes telles que :



- Les colonnes qui sont contient des valeurs réelsn nous avaons utilisées la moyenne pour les remplires.
- Les colonnes qui sont contient des modalités, nous avaons utilisées le mode pour les remplires.

- la moyenne

```python
colonnes_specifiees = ['host_listings_count', 'host_total_listings_count', 
                        'review_scores_rating', 'reviews_per_month']

for colonne in colonnes_specifiees:
    moyenne_colonne = boston[colonne].mean()
    boston[colonne].fillna(moyenne_colonne, inplace=True)
```

- le mode

```python
colonnes_specifiees = ['bathrooms', 'bedrooms', 'review_scores_accuracy', 'review_scores_cleanliness', 
                       'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 
                       'review_scores_value', "beds"]

for colonne in colonnes_specifiees:
    mode = boston[colonne].value_counts().idxmax()
    boston[colonne].fillna(mode, inplace=True)
```

```python
boston.isnull().sum()
```

## <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;padding:5px">correlation matrix</p>


- Concernant les colonnes numériques,
- Nous examinerons leurs corrélations et supprimerons l'une d'entre elles en cas de corrélation élevée entre elles.

```python
column_types = boston.dtypes
numeric_columns = column_types[column_types != 'object'].index
correlation_matrix = boston[numeric_columns].corr()
correlation_matrix.T.head(3)
```

Liste pour stocker les paires de colonnes fortement corrélées

```python
seuil_correlation = 0.85

colonnes_a_supprimer = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > seuil_correlation:
            colonnes_a_supprimer.append((correlation_matrix.columns[i], correlation_matrix.columns[j]))
```

Afficher les paires de colonnes fortement corrélées

```python
print("\nColonnes fortement corrélées :")
for colonne1, colonne2 in colonnes_a_supprimer:
    print(f"{colonne1} - {colonne2} : {correlation_matrix.loc[colonne1, colonne2]}")
```

- Supprimer l'une des colonnes de chaque paire fortement corrélée :
- On va Supprimer la colonne avec la plus faible variance

```python
for colonne1, colonne2 in colonnes_a_supprimer:    
    if colonne1 in boston.columns and colonne2 in boston.columns:
        if boston[colonne1].var() > boston[colonne2].var():
            boston.drop(colonne2, axis=1, inplace=True)
        else:
            boston.drop(colonne1, axis=1, inplace=True)
boston.head(3) 
```

## <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;padding:7px">Data Scaling</p>

```python
numeric_columns = boston.select_dtypes(include=['float64', 'int64']).columns
standard_scaler = StandardScaler()
boston[numeric_columns] = standard_scaler.fit_transform(boston[numeric_columns])
boston.head(3)
```

```python
print("- La taille finale de notre dataset :", boston.shape)
```

- Voila, notre dataset maintenant a été bien nettoyée.
