```python
import requests

def get_video_comments(video_id, api_key):
    base_url = "https://www.googleapis.com/youtube/v3/commentThreads"
    params = {
        "part": "snippet",
        "videoId": video_id,
        "key": api_key,
        "maxResults": 800,  # Set maxResults to 100 per page (maximum allowed)
    }

    comments = []

    # Make initial request to get comments
    response = requests.get(base_url, params=params)
    data = response.json()

    # Process the first page of comments
    while data.get("items"):
        for item in data["items"]:
            comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
            comments.append(comment)

        # Check if there is another page of comments
        if "nextPageToken" in data:
            params["pageToken"] = data["nextPageToken"]  # Set the next page token
            response = requests.get(base_url, params=params)
            data = response.json()
        else:
            break  # No more pages, exit the loop

    return comments


# Usage: Replace 'video_id' and 'api_key' with your specific video ID and API key

video_id = "BJ2VqM1V_hQ"
api_key = "AIzaSyDBN5uCPBlFCxQgomYUW43_h32Xu1dIFpI"
video_comments = get_video_comments(video_id, api_key)

# Print all comments retrieved
for comment in video_comments:
    print(comment)
```

**- Fonction pour crÃ©er les n-grammes Ã  partir des phrases tokenisÃ©es**

```python
from nltk.util import ngrams
from collections import defaultdict
import random

# TÃ©lÃ©chargement des ressources nÃ©cessaires pour nltk
nltk.download('punkt')

def create_ngrams(phrase_propre, n):
    ngram_counts = defaultdict(int)
    for sentence in phrase_propre:
        words = sentence.split()  # SÃ©parer la phrase en mots
        for i in range(len(words) - n + 1):
            ngram = tuple(words[i:i+n])  # CrÃ©er le n-gramme comme un tuple de mots
            ngram_counts[ngram] += 1
    return ngram_counts
```

**- Fonction pour prÃ©dire le mot suivant en fonction des n-1 mots prÃ©cÃ©dents**

```python
def predict_next_word(previous_words, ngram_counts):
    next_word_candidates = []
    previous_words_tuple = tuple(previous_words)
    for ngram, count in ngram_counts.items():
        if ngram[:-1] == previous_words_tuple:
            next_word_candidates.append((ngram[-1], count))
    if not next_word_candidates:
        return None
    else:
        total_count = sum(count for _, count in next_word_candidates)
        next_word_probs = [(word, count / total_count) for word, count in next_word_candidates]
        return random.choices([word for word, _ in next_word_probs], [prob for _, prob in next_word_probs])[0]
```

Exemple d'utilisation :

```python
# CrÃ©ation des n-grammes
n = 3  # Choisir la taille du n-gramme (par exemple, 3 pour les trigrammes)
ngram_counts = create_ngrams(phrase_propre, n)
```
