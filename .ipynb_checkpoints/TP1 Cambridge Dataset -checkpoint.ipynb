## <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">Dropping columns that are of no use</p>


### Suppression selon le nombre des valeurs nulles : 


- Les colonnes contenant plus de 70 % de valeurs manquantes seront supprimées.

```python
columns_to_drop = []
for column in data.columns:
    null_count = data[column].isna().sum()
    total_rows = data.shape[0]
    if (null_count / total_rows) >= 0.7:
        columns_to_drop.append(column)
        print(f"{column}: {null_count}")
```

```python
data.drop(columns=columns_to_drop, axis=1, inplace=True)
```

### - Suppression selon le nombre des enregitrements dupliquées :

```python
nbr_duplicate = data.duplicated().sum()
print("- Le nombre des doublons dans notre dataset est: ", nbr_duplicate)
print("- Le nombre des colonnes restent :", len(data.columns))
```

### - Suppression des villes qui n'appartiennent pas à Cambridge

```python
cities_to_remove = ['Somerville', 'Watertown', 'Arlington', 
                    'Belmont', 'Boston', 'Brookline' ]

data = data[~data['city'].isin(cities_to_remove)]
```

### - Suppression des identifiants et des données ont une relation avec celles récupérées

```python
data.drop(columns=['id', 'scrape_id', 'last_scraped', 'host_id'], axis=1, inplace=True)
```

<a id="1"></a>
## <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">Data Transform</p>


### - Nous allons transférer 't' à 1 et 'f' à 0 dans ces colonnes

```python
columns_to_transform = ['require_guest_profile_picture', 'require_guest_phone_verification', 'requires_license', 'instant_bookable',   
                        'is_business_travel_ready', 'has_availability', 'is_location_exact', 'host_has_profile_pic', 
                        'host_identity_verified']

mapping = {'t': 1, 'f': 0, 'nan': 0, np.nan: 0}

data[columns_to_transform] = data[columns_to_transform].replace(mapping)
```

### - Convertir la colonne 'price' 'extra_people' 'cleaning_fee' 'security_deposit' en float

```python
# Remove non-numeric characters ('$' and ',') from the values in the columns
data['price'] = data['price'].str.replace('[\$,]', '', regex=True)
data['extra_people'] = data['extra_people'].str.replace('[\$,]', '', regex=True)
data['cleaning_fee'] = data['cleaning_fee'].str.replace('[\$,]', '', regex=True)
data['security_deposit'] = data['security_deposit'].str.replace('[\$,]', '', regex=True)

# Convert the values in the columns to float
data['price'] = data['price'].astype(float)
data['extra_people'] = data['extra_people'].astype(float)
data['cleaning_fee'] = data['cleaning_fee'].astype(float)
data['security_deposit'] = data['security_deposit'].astype(float)
```

### - Suppression toutes les colonnes de type object

```python
object_column = []
for column in data.columns :
    if data[column].dtype == 'object' :
        object_column.append(column)
object_column
```

```python
data.drop(columns=object_column, axis=1, inplace=True)
```

<a id="1"></a>
## <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">Filling missing values</p>


- La moyenne

```python
colonnes_specifiees = ['host_listings_count', 'host_total_listings_count', 'reviews_per_month']

for colonne in colonnes_specifiees:
    moyenne_colonne = data[colonne].mean()
    data[colonne].fillna(moyenne_colonne, inplace=True)
```

- Le mode

```python
colonnes_specifiees = ['beds', 'security_deposit', 'cleaning_fee', 'review_scores_rating', 'review_scores_accuracy',  
                       'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication',  
                       'review_scores_location', 'review_scores_value']

for colonne in colonnes_specifiees:
    mode = data[colonne].value_counts().idxmax()
    data[colonne].fillna(mode, inplace=True)
```

## <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">correlation matrix</p>


```python
matrix_corr = data.corr()
matrix_corr
```

- Les colonnes qui ont une forte corrélation entre elles

```python
seuil_corr = 0.85

paires_corr = []


for i in range(len(matrix_corr.columns)):
    for j in range(i+1, len(matrix_corr.columns)):  # Éviter de comparer une colonne à elle-même
        if abs(matrix_corr.iloc[i, j]) >= seuil_corr:  # Vérifier si la corrélation est supérieure ou égale au seuil
            paires_corr.append((matrix_corr.columns[i], matrix_corr.columns[j]))

for paire in paires_corr:
    print("{} et {} : {}.".format(paire[0], paire[1], matrix_corr.loc[paire[0], paire[1]]))
```
